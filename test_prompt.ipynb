### Prompt design

````python
openai_client = OpenAI(api_key="OPENAI_KEY")

def call_gpt4o_mini(message):
    """Call the GPT model for text information and return the response."""
    try:
        response = openai_client.chat.completions.create(
            model = "gpt-4o-mini",
            messages=[{"role": "user", 
                       "content": message}],
            temperature=0.0,
            max_tokens=8000
        )
        return response.choices[0].message.content
    except Exception as e:
        print(f"Error calling GPT: {e}")
        return None
````

````python
def create_prompt(lead, company, candidates):
    prompt = f"""
    As the lead investment firm, {lead['name']}, your goal is to select three out of the five candidate firms to collaborate on an investment \
    opportunity with {company['name']}, a {company['stage']} company specializing in {company['industry']}.

    Here are some key considerations to guide your decision-making:
    1. {company['name']} is located in {company['location']} and operates in the {company['industry']} industry.
    2. {company['name']} aligns with your focus on media and technology sectors and matches your stage preference ({lead['stage_preference']}).
    3. Your geographical focus is {lead['geo_focus']}.

    Below are the candidates:
    """
    
    for idx, candidate in enumerate(candidates, start=1):
        prompt += f"\n{idx}. {candidate['name']} ({candidate['location']}): {candidate['description']} Focus: {candidate['focus']}, " \
                  f"Stage: {candidate['stage']}, AUM: {candidate['aum']}"
    
    prompt += """
    \nBased on their focus, stage alignment, geographical location, and other strategic factors, please select three candidates \
    and provide reasoning for each choice.

    Output your result in JSON format. 
    {
        Candidates: <Candidate 1, Candidate 2, Candidate 3>,
        Reasoning: <Your Reasoning>
    }
    """
    return prompt
````
